```python
t=np.arange(0.,5.,0.2)
plt.figure()
plt.subplot(1,3,1)
plt.plot(t,t,'r--')
plt.subplot(1,3,2)
plt.plot(t,t**2,'bs')
plt.subplot(1,3,3)
plt.plot(t,t**3,'g^')
plt.tight_layout()
plt.show()
```

```python
t=np.arange(0.,5.,0.2)
plt.figure()
plt.subplot(3,1,1)
plt.plot(t,t,'r--')
plt.subplot(3,1,2)
plt.plot(t,t**2,'bs')
plt.subplot(3,1,3)
plt.plot(t,t**3,'g^')
plt.tight_layout()
plt.show()
```

```python
t=np.arange(0.,5.,0.2)
plt.figure()
plt.subplot(1,3,1)
plt.plot(t,t,'r--')
plt.title("Linear Subplot")
plt.xlabel('t values')
plt.ylabel('output labels')

plt.subplot(1,3,2)
plt.plot(t,t**2,'bs')
plt.title("Quadratic Subplot")
plt.xlabel('t values')
plt.ylabel('output labels')

plt.subplot(1,3,3)
plt.plot(t,t**3,'g^')
plt.title("Cubic Subplot")
plt.xlabel('t values')
plt.ylabel('output labels')

plt.suptitle("Differetn Functions of t")
plt.tight_layout()
plt.show()
```

```python
x=np.linspace(1,10,100)
fig,axes=plt.subplots(2,3)
axes[0,0].plot(x,np.sin(x))
axes[0,0].set_title('sin func')
axes[0,0].set_xlabel('x')
axes[0,0].set_ylabel('y')


axes[0,1].plot(x,np.cos(x))
axes[0,1].set_title('cos func')
axes[0,1].set_xlabel('x')
axes[0,1].set_ylabel('y')

axes[0,2].plot(x,np.tan(x))
axes[0,2].set_title('tan func')
axes[0,2].set_xlabel('x')
axes[0,2].set_ylabel('y')

axes[1,0].plot(x,np.exp(x))
axes[1,0].set_title('exp func')
axes[1,0].set_xlabel('x')
axes[1,0].set_ylabel('y')

axes[1,1].plot(x,np.sqrt(x))
axes[1,1].set_title('sqrt func')
axes[1,1].set_xlabel('x')
axes[1,1].set_ylabel('y')

axes[1,2].plot(x,np.log(x))
axes[1,2].set_title('log func')
axes[1,2].set_xlabel('x')
axes[1,2].set_ylabel('y')
plt.tight_layout()
```

# scatter plots

```python
x=np.random.rand(100)*10
y=np.random.rand(100)*10
plt.scatter(x,y,marker='p',edgecolor='r',linewidth=2,alpha=0.5,s=50,label='random')
plt.legend()
plt.grid(True)
plt.show()
```

### bar plot

```python
courses=['genAI','HLD','HHLD','LLD','DSA','CPP']
students=[2500,3000,3200,2000,2700,1000]
hatches=['/','*','o','\\','|','-']
bar_colors=['red','blue','magenta','yellow','orange','green']
bars=plt.bar(courses,students,width=0.5,align='center',edgecolor='black',linewidth=3)
for bar ,hatch,color in zip(bars,hatches,bar_colors):
    bar.set_hatch(hatch)
    bar.set_color(color)
    bar.set_edgecolor('black')
plt.xlabel('courses')
plt.ylabel('students')
plt.title("no of students enrolled in courses")
plt.show()
```

### Histogram

```python
ages=[34,23,78,54,67,45,21,98,54,76]
sorted_ages=np.sort(ages)
plt.hist(ages,bins=5,orientation='horizontal',color='r',edgecolor='black')
plt.xlabel('Frequency')
plt.ylabel('ages')
plt.title('Ages Plot')
plt.show()
```

### Pie Plot

```python
courses=['genAI','HLD','HHLD','LLD','DSA','CPP']
students=[2500,3000,3200,2800,2700,1000]
explode=[0.2,0,0,0,0,0]
pie_colors=['red','blue','magenta','yellow','orange','green']
plt.pie(students,labels=courses,shadow=True,explode=explode,colors=pie_colors,autopct='%.2f%%')
plt.legend(loc='lower right')
plt.show()
```

### 3D Scatter plot

```python
ax=plt.axes(projection ='3d')
x=np.random.rand(100)
y=np.random.rand(100)
z=np.random.rand(100)
ax.scatter(x,y,z)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```

# 3D  plot

```python
ax=plt.axes(projection ='3d')
x=np.arange(0,50,0.1)
y=np.sin(x)
z=np.cos(x)
ax.plot(x,y,z)
plt.show()
```

### 3D bar Plots

```python
ax=plt.axes(projection='3d')
x=np.random.randint(1,10,5)
y=np.random.randint(1,10,5)
z=np.zeros(5)
dx=np.ones(5)
dy=np.ones(5)
dz=np.random.randint(1,10,5)
ax.bar3d(x,y,z,dx,dy,dz)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```
